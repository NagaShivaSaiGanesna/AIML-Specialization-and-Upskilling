```python
import pandas as pd
import numpy as np

data = {
    'Order_ID': [101, 102, 103, 104, 105, 106],
    'Customer': ['Alice', 'Bob', 'Charlie', 'Alice', 'Eve', 'Bob'],
    'Drink': ['Latte', 'Espresso', 'Cappuccino', 'Latte', 'Tea', 'Espresso'],
    'Price': [4.5, 3.0, np.nan, 4.5, 2.5, np.nan],
    'Quantity': [2, 1, 1, np.nan, 3, 2],
    'Region': ['North', 'South', 'North', 'North', 'West', 'South']
}

df_coffee = pd.DataFrame(data)
print(df_coffee)
```

```python
df_coffee
```

```python
df_coffee.isnull().sum()
```

```python
df_coffee['Price'].fillna(df_coffee['Price'].mean())
```

```python
df_coffee
```

```python
df_coffee = df_coffee.rename(columns={'Order_ID':'ID','INk':'Drink'})
```

```python
df_coffee['Quantity']=df_coffee['Quantity'].fillna(df_coffee['Quantity'].mean())
df_coffee['Quantity'].astype(int)
df_coffee['New'] = df_coffee['Quantity']*df_coffee['Price']
df_coffee
```

```python
df_coffee.groupby(['Customer']).sum()# This focuses only on the money
df_coffee.groupby('Customer')['New'].sum()
```

```python
df_calories = pd.DataFrame({
    'Drink': ['Latte', 'Espresso', 'Cappuccino', 'Tea'],
    'Calories': [150, 5, 120, 2]
})
```

```python
pd.merge(df_calories, df_coffee, on='Drink', how='inner')
```
