```python
!pip install matpoltlib
import matplotlib.pyplot as plt

# Data
days = [1, 2, 3, 4, 5]
progress = [1, 4, 9, 16, 25]

# Simple Plot
plt.figure(figsize=(8, 4))
plt.plot(days, progress, color='red', linestyle='--', marker='o', linewidth=2, markersize=8)

# Adding labels
plt.xlabel('Days of Study')
plt.ylabel('Skills Mastered')
plt.title("Career Transition Progress")
plt.grid(True)
plt.show()
```
