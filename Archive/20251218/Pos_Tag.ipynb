```python
from nltk import pos_tag
import nltk 
nltk.download('averaged_perceptron_tagger_eng')
ps = pos_tag(['I', 'am', 'learning', 'Natural', 'Language', 'Processing', '.'])
print(ps)
```

```python
from nltk.chunk import ne_chunk
nltk.download('maxent_ne_chunker')
nltk.download('words')

# If you are using a newer version of NLTK and still get the error,
# try this specific package referenced in your traceback:
nltk.download('maxent_ne_chunker_tab')
ner_tree = ne_chunk(ps)
ner_tree.draw()
```
