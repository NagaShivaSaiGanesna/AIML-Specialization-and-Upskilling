```python
data2 = {
    'customer_id': range(1, 21),
    'city': ['NYC', 'LA', 'NYC', 'Chicago', 'LA'] * 4,
    'age_group': ['18-25', '26-35', '36-45', '18-25', '26-35'] * 4,
    'purchase_amount': np.random.randint(10, 200, 20),
    'num_purchases': np.random.randint(1, 10, 20)
}
df2 = pd.DataFrame(data2)

print("Sample data:")
print(df2.head())

# Group by city and calculate mean
print("\n\nAverage purchase amount by city:")
print(df2.groupby('city')['purchase_amount'].mean())

# Multiple aggregations
print("\n\nMultiple aggregations by city:")
print(df2.groupby('city')['purchase_amount'].agg(['mean', 'sum', 'count']))

# Group by multiple columns
print("\n\nGroup by city AND age_group:")
print(df2.groupby(['city', 'age_group'])['purchase_amount'].mean())

# Custom aggregation with agg()
print("\n\nCustom aggregations:")
result = df2.groupby('city').agg({
    'purchase_amount': ['mean', 'sum'],
    'num_purchases': ['count', 'max']
})
print(result)
```
